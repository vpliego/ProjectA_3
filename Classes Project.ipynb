# Brief Description of class Datawork 

This class works with the ncbirths dataset, which is a dataset that collected birth information on births in North Carolina in 2004. This dataset is a random sample of 1000 cases, and has 13 variables ranging from birth weight to the mothers' smoking status (whether they smoke or not). The purpose of this dataset is to determine a relationship (if any) between habits of expectant mothers and the birth of their children. Therefore, this class aims to work with the dataset by doing basic data wrangling to clean up the dataset (removing NA values), compute the mean of a variable for statistical analysis, and finally plot two variables of interest. 

# List of class internal attributes + description 

**file_path:** This internal attribute will allow my class to read a csv file as the file_path will be unique to each user and their computer, therefore, it is an internal attribute in this class. 

**var_x:** This internal attribute will allow my class to work with a variable "x" of interest from the ncbirths dataset which will become useful during the statistical analysis portion, as well as the data visualization portion of the class. 

**var_y:** Similar to var_x, this internal attribute will allow my class to work with a variable "y" of interest from the ncbirths dataset which will become useful during the data visualization portion of the class.

**update_df:** update_df will ensure that the user is working with the updated version of the dataframe, after the data wrangling has been applied to the dataset, to ensure the user is plotting data that will not have missing data values, as well as calculating any stats that are missing said values. 

**df:** df will allow the dataset to be read in the function read_data. 

**mean_var_x:** This will allow the user to calculate the mean of variable x in the stat_analysis function. 

**mean_var_y:** Similarly, this will allow the user to calculate the mean of variable y in the stat_analysis function. 



# List of class internal functions + description 

**__init__():** This is the constructor function and will initialize the class when an instance is created.  

**read_data():** This function will take in the file_path internal attribute and will read the csv file provided. 

**data_wrangling():** This function will take the dataframe and drop any values where they are NA, and then will be stored in update_df so that this can be used instead of the original dataframe. It will also even confirm to the user that there are no NA values present in the dataframe by checking if this is true using "isnull().values.any()" function. 

**stat_analysis():** This function will print the mean of both x and y variables of interest. 

**data_exploratory():** This function will simply return a pairplot for the user to better understand relationships between the x and y variables of interest. 

**data_visualization():** This function will be more specific in plotting the relationship of the x and y variables of interest using a scatterplot. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class Datawork():
    def __init__(self, file_path, var_x, var_y): 
        self.file_path = file_path
        self.var_x = var_x
        self.var_y =  var_y
        self.update_df = None
        self.df = self.read_data()
        self.mean_var_x = None
        self.mean_var_y = None 
        
      
    def read_data(self):
        df = pd.read_csv(self.file_path)
        return df
    
    def data_wrangling(self):
        self.update_df = self.df.dropna()
        return f'Are there any n/a values in this dataframe (T/F): {self.update_df.isnull().values.any()}'
         
    def stat_analysis(self):
        self.mean_var_x = self.update_df[self.var_x].mean()
        self.mean_var_y = self.update_df[self.var_y].mean()
        
        mean_x = self.mean_var_x
        mean_y = self.mean_var_y 
        print(f'The mean of {var_x} is: {mean_x}') 
        print(f'The mean of {var_y} is: {mean_y}')
              
    def data_exploratory(self): #this is a starter for visualizing 
        pairplot = sns.pairplot(data=self.update_df)
        return pairplot
    def data_visualization(self): 
        plt.scatter(self.update_df[self.var_x], self.update_df[self.var_y])
        plt.title(f'The relationship between {self.var_x} and {self.var_y}')
        plt.xlabel(self.var_x)
        plt.ylabel(self.var_y)
        plt.show()
        
        
        
    
    
```
