```python
new_books_data = books_data.fillna(" ")
new_books_data.pop("title")
new_books_data.pop("ratingHistogram")
new_books_data
```

```python
type(new_books_data["genres"].iloc[0])
```

```python
new_books_data.loc[:,"genres"].apply(eval)
```

```python
new_books_data.loc[:,"genres"] = new_books_data.loc[:,"genres"].apply(eval)
new_books_data["genres"] 
```

```python
new_books_data.loc[:,"num_genres"] = new_books_data.apply(lambda x: len(x['genres']),axis = 1)
```

```python
new_books_data['ratingsCount'] = pd.to_numeric(new_books_data['ratingsCount'], errors = 'coerce')
new_books_data['reviewsCount'] = pd.to_numeric(new_books_data['reviewsCount'], errors = 'coerce')
new_books_data['numPages'] = pd.to_numeric(new_books_data['numPages'], errors = 'coerce')
```

```python
book_data_genre = new_books_data.explode('genres').drop_duplicates()
```

```python
mask_genres = (book_data_genre['genres'] == "Young Adult Contemporary") | (book_data_genre['genres'] == "Adventure") | (book_data_genre['genres'] == "Supernatural")
df_masked = book_data_genre[mask_genres]

sns.relplot(data = df_masked, x = 'ratingsCount', y = 'numPages', hue = 'genres')
```

Before, we looked at the relationship that ratings count and reviews count had, however here we are interested in how ratings count is affected by the number of pages of a book. However, adventure, when it doesn't have such a large number of pages (between 0-2,000) tends to have the most ratings given.
