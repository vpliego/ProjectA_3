### Instance methods (functions)

Let's add two instance methods (internal functions) to this class.

```python
#change name
#change major
#describe/display
#calculating years to graduation
```

```python
class Smithie:
    school = "Smith College"
    def __init__(self, name, year):
        self.name = name
        self.year = year
        self.major = "Undecided"
    
    def describe(self):
        return f"{self.name} is a {self.school} student whose major is {self.major} and is graduating in {self.year}"
    
    def change_major(self, new_major):
        if self.major == "Undecided":
            print(f"{self.name} is delcaring her major!!")
        else:
            print(f"{self.name} is changing her major!!")
        self.major = new_major
        
    def years_to_graduation(self):
        curr_years = 2023
        years_left = self.year - curr_year
        return years_left
```

```python
class Resident(Smithie):
    neighborhood = None
    house_name = None
    
    def set_house(self, house_name, neighborhood):
        self.house_name = house_name
        self.neighborhood = neighborhood
        
    def lives_in(self):
        if self.house_name == None:
            return f"{self.name} doesn't have a house"
        else:
            return f"{self.name} lives in {self.house_name} House"
```

```python
Casey = Resident("Casey", 2025)
```

```python
class restaurant:
    location = "Northampton" #adding default location
    
    def __init__(self, name, food_type):
        self.name = name #takes in a value for restaurant name 
        self.food_type = food_type #takes in the type of cuisine 
        
        
    
    def describe(self): #adding basic method that will describe where the user is (restaurant) and their default location set above
        return f"Hello! You are at {self.name} in {self.location}. Welcome in!"
    
    def get_menu(self): #method will print out a menu according to different cuisine types
        if self.food_type == "Italian":
            return ["Pesto Pasta: $19.99", "Penne Bolognese: $27", "Penne Alfredo: $27", 'Shrimp Cocktail: $14']
        elif self.food_type == 'Mexican':
            return ["Chiles Rellenos de queso (stuffed peppers with cheese): $15.95", "Beef Burrito: $10.95", "Chicken Burrito: $10.95", "Shrimp Taco: $3.95"]
        elif self.food_type == "American":
            return ["Burger: $13", "Fries: $6.95", "Cheese Fries: $7.95", "Bacon Burbon Burger: $17.95", "Club Sandwich: $11.95", "Grilled Cheese: $9.95"]
        else: #if user does not input one of the 3 cuisines provided above, then this message will display 
            return "Sorry! There are currently no menu options available to display for this restaurant!"  
```

```python
#using campus cafe as first example 
campus = restaurant("Campus Cafe", "American")
campus.describe()
```

```python
campus.get_menu()
```

```python
#using taco bell as second example
taco_bell = restaurant("Taco Bell", "Mexican")
taco_bell.describe()
```

```python
taco_bell.get_menu()
```

```python
#using india house as third example, the describe() will work because this is a basic method that simply describes where you are. 
indian = restaurant("India House", "Indian")
```

```python
#instance where the get_menu() goes to the 'else' statement since "indian" isn't one of the listed cuisines. 
indian.get_menu()
```
